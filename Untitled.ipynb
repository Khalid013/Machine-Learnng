```python
a=df['max_life_span'].mean()
b=df['gestation_time'].mean()
c=df['total_sleep'].mean()
print(f"a={a} , b={b} and c={c}")
```

```python
df['max_life_span']=df['max_life_span'].fillna(a)
df['gestation_time']=df['gestation_time'].fillna(b)
df['total_sleep']=df['total_sleep'].fillna(c)
```
