```python
cig=df['cigsPerDay'].mean()
cig
```

```python
BPM=df['BPMeds'].median()
BPM
```

```python
Chol=df['totChol'].median()
Chol
```

```python
BM=df['BMI'].median()
BM
```

```python
Heart=df['heartRate'].median()
Heart
```

```python
Gloco=df['glucose'].median()
Gloco
```

```python
df['cigsPerDay']=df['cigsPerDay'].fillna(cig)
df['BPMeds']=df['BPMeds'].fillna(BPM)
df['totChol']=df['totChol'].fillna(Chol)
df['BMI']=df['BMI'].fillna(BM)
df['heartRate']=df['heartRate'].fillna(Heart)
df['glucose']=df['glucose'].fillna(Gloco)
```

```python
df.isnull().sum()
```

```python
df
```

```python
x=df.drop(['TenYearCHD'],axis=1)
x
```

```python
y=df['TenYearCHD']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
x_train
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```
